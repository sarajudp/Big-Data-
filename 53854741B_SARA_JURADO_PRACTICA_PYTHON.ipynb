```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(
    r'datasets/Info_pais.csv',
    encoding='ISO-8859-1',
    delimiter=';'
)
```

```python
df.head()
```

```python
df_order = df.sort_values(
    'Esperanza de vida',
    ascending=True
)
```

```python
df_order.head()
```

```python
plt.scatter(df_order['Renta per capita'], df_order['Esperanza de vida']);
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida']
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida');
```

```python
df_order['Poblacion_normalizada'] = df_order['Poblacion'] / max(df_order['Poblacion'])
```

```python
df_order.head()
```

```python
df_order['Poblacion_normalizada'] = df_order['Poblacion'] / (max(df_order['Poblacion']) / 10000)
```

```python
df_order.head()
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada']
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida');
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada']
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5, 10);
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada']
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5, 10);
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada']
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5, 10);

for i in range(0, 10):
    plt.annotate(
        df_order['País'][i],
        (
            df_order['Renta per capita'][i],
            df_order['Esperanza de vida'][i]
        )
    )
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada']
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5, 10)

for i in range(0, 10):
    plt.annotate(
        df_order['País'][i],
        (
            df_order['Renta per capita'][i],
            df_order['Esperanza de vida'][i]
        )
    )

plt.yticks(ticks=range(1, 120, 10));
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada'],
    alpha=0.5
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5, 10)

for i in range(0, 10):
    plt.annotate(
        df_order['País'][i],
        (
            df_order['Renta per capita'][i],
            df_order['Esperanza de vida'][i]
        )
    )

plt.yticks(ticks=range(1, 120, 10));
```

El gráfico de dispersión muestra la relación entre la renta per cápita y la esperanza de vida en diferentes países. En el rango más bajo de renta per cápita (de 0 a 20000), los países muestran una gran variación en esperanza de vida (entre 53.7 y 82.2). A partir de este punto, observamos una relación positiva entre las variables, es decir, a más renta per cápita, mayor la esperanza de vida. Esta relación se vuelve menos evidente pasado el rango de los 30000 de renta per cápita, ya que la esperanza de vida no aumenta significativamente, e incluso disminuye en países como Estados Unidos, a pesar del aumento en las rentas. Por otro lado, el tamaño de los puntos representa la población normalizada de cada país, y no parece tener una relación clara con la renta per cápita o la esperanza de vida, aunque los dos países más grandes (India y China) coinciden en el rango más bajo de renta per cápita. 

Por lo tanto, la renta per cápita influye positivamente en la esperanza de vida, pero no parece ser el único factor determinante. Otros aspectos como la calidad del aire y del agua, los conflictos políticos, la desigualdad económica, la incidencia de enfermedades, la sanidad, educación, las condiciones laborales o el acceso a la vivienda, entre otros, también influyen en la esperanza de vida y podrían explicar la variación en aquellos puntos en los que la renta per cápita no es completamente determinante. Por ello, para un análisis más completo, deberían valorarse también estos factores.
